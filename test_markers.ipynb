```python
import pandas as pd
```

```python
df = pd.read_csv("AppleGenotypes.csv", index_col="code")
df
```

```python
df = df.replace("AB", 1).replace("A", 0).replace("B", 2)
df
```

```python
def clean(x):
    try:
        x = int(x)
        if x >= 0 and x <= 2:
            return x
    except Exception:
        pass

    return "x"

df = df.applymap(
    lambda x: clean(x)
)
df
```

```python
nrows = df.shape[0]
```

```python
for i in range(0, nrows):
    alleles = {}
    fails = 0

    row = df.iloc[i]
    counts = row.value_counts()
    
    for key in counts.keys():
        if key == "x":
            fails = int(counts["x"])
        else:
            alleles[key] = int(counts[key])

    n_alleles = len(alleles.keys())

    rowlen = len(row)

    callrate = float(rowlen - fails) / rowlen

    print(f"Number of fails == {fails}")
    print(f"Alleles == {alleles}")
    print(f"n_alleles == {n_alleles}")
    print(f"call {callrate} : {rowlen}")
```
