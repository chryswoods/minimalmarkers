```python
import pandas as pd
```

```python
df = pd.read_csv("AppleGenotypes.csv", index_col="code")
df
```

```python
df = df.replace("AB", 1).replace("A", 0).replace("B", 2)
df
```

```python
def clean(x):
    try:
        x = int(x)
        if x >= 0 and x <= 2:
            return x
    except Exception:
        pass

    return "x"

df = df.applymap(
    lambda x: clean(x)
)
df
```
